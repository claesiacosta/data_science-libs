```python
import seaborn as sns
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(data=tips, palette='rainbow', orient='h')
```

```python
sns.violinplot(x='day', y='tip', data=tips)
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(data=tips, y='total_bill', x='day')
```

```python
sns.stripplot(data=tips, y='total_bill', x='day', jitter=True)
```

```python
sns.stripplot(data=tips, y='total_bill', x='day', jitter=True, hue='sex')
```

```python
sns.stripplot(data=tips, y='total_bill', x='day', jitter=True, hue='sex', dodge=True)# ==  split=True
```

```python
sns.swarmplot(data=tips, y='total_bill', x='day')
```

```python
sns.swarmplot(data=tips, y='total_bill', x='day', hue='sex')
```

```python
sns.swarmplot(data=tips, y='total_bill', x='day', hue='sex', split=True)
```

```python
sns.swarmplot(data=tips, y='total_bill', x='day', color='black')
sns.violinplot(data=tips, y='total_bill', x='day')
```

```python
sns.factorplot(x='sex', y='total_bill', data=tips, kind='bar')
```

```python
sns.factorplot(x='sex', y='total_bill', data=tips, kind='violin')
```

```python
sns.factorplot(x='sex', y='total_bill', data=tips, kind='swarm')
```
