```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x*x
```

```python
y
```

```python
#Funcional
#plt.plot(x,y,'r')#colore a linha
plt.plot(x,y, 'r--')#tracejado
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
fig= plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])

axes1.plot(x,y,'r')
axes1.set_xlabel('Eixo X')
axes1.set_title('Título')
axes2.plot(y,x,'b')
```

```python
fig, ax = plt.subplots()
ax.plot(x,x**3,'b--')
ax.plot(x,x**4,'g-.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Título')
```

```python
fig, axes = plt.subplots(figsize=(8,8))

axes.plot(x*1.5,y*1.5,'r--', linewidth=2)
axes.plot(x,y*2,'b-.', linewidth=2)
axes.plot(x*2,y,'g*-', linewidth=2)
axes.set_xlabel("Eixo X")
axes.set_ylabel("Eixo Y")
axes.set_title('Título')
```

```python
fig, axes = plt.subplots(figsize=(8,8))

axes.plot(x,y,'r',label='Dados 1')
axes.plot(y,x,'b-.', label='Dados 2')
axes.set_xlabel("Eixo X")
axes.set_ylabel("Eixo Y")
axes.set_title('Título')
axes.legend(loc=0)
```
